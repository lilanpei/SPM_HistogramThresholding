```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import  pylab
```

```python
threads = np.loadtxt("Threads.txt")
farm = np.loadtxt("Farm.txt")
farm_pipeline = np.loadtxt("farm_pipeline.txt")
sequential = 621052
```

```python
s_threads = sequential/threads
s_farm = sequential/farm
s_farm_pipeline = sequential/farm_pipeline
idea = [3*i+1 for i in range(100)]
plt.figure(figsize=(20,10))
plt.xlabel('nWorkers')
plt.ylabel('Tseq/Tpar(n)')
plt.title('SpeedUp')
plt.xticks([i for i in range(100)],[3*i+1 for i in range(100)],rotation=90)
plt.axis([0, 90, 0, 22])
plt.plot(s_threads)
plt.plot(s_farm)
plt.plot(s_farm_pipeline)
plt.plot(idea)
plt.legend(('C++Threads', 'Farm', 'Pipeline embedded in Farm', 'idea'), loc='best')
plt.show()
```

```python
scalab_threads = threads[0]/threads
scalab_farm = farm[0]/farm
scalab_farm_pipeline = farm_pipeline[0]/farm_pipeline
idea = [3*i+1 for i in range(100)]
plt.figure(figsize=(20,10))
plt.xlabel('nWorkers')
plt.ylabel('Tpar(1)/Tpar(n)')
plt.title('Scalability')
plt.xticks([i for i in range(100)],[3*i+1 for i in range(100)],rotation=90)
plt.axis([0, 99, 0, 22])
plt.plot(scalab_threads)
plt.plot(scalab_farm)
plt.plot(scalab_farm_pipeline)
plt.plot(idea)
plt.legend(('C++Threads', 'Farm', 'Pipeline embedded in Farm', 'idea'), loc='best')
plt.show()
```

```python
n = [3*i+1 for i in range(100)]
n_farm_pipeline = [6*i+1 for i in range(85)]
e_threads = sequential/(n*threads)
e_farm = sequential/(n*farm)
e_farm_pipeline = sequential/(n_farm_pipeline*farm_pipeline)
idea = np.ones(100)
plt.figure(figsize=(20,10))
plt.xlabel('nWorkers')
plt.ylabel('Tseq/(n*Tpar(n))')
plt.title('Efficiency')
plt.xticks([i for i in range(100)],[3*i+1 for i in range(100)],rotation=90)
plt.axis([0, 99, 0, 1.6])
plt.plot(e_threads)
plt.plot(e_farm)
plt.plot(e_farm_pipeline)
plt.plot(idea)
plt.legend(('C++Threads', 'Farm', 'Pipeline embedded in Farm', 'idea'), loc='best')
plt.show()
```
